```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome('C:/Users/nextop/Downloads/chromedriver_win32/chromedriver.exe')
driver.implicitly_wait(3)
driver.get("https://www.livesport.com/team/manchester-united/ppjDR086/")
page=driver.page_source
bs_obj=BeautifulSoup(page,"html.parser")
win =bs_obj.find_all("span",{"class":"formIcon formIcon--w"})
tiewin =bs_obj.find_all("span",{"class":"formIcon formIcon--wo"})
draw = bs_obj.find_all("span",{"class":"formIcon formIcon--d"})
lose = bs_obj.find_all("span",{"class":"formIcon formIcon--l"})
tielose = bs_obj.find_all("span",{"class":"formIcon formIcon--lo"})
n_win =len(win)+len(tiewin)
n_draw=len(draw)
n_lose = len(lose)+len(tielose)
print(n_win)
print(n_draw)
print(n_lose)
```
